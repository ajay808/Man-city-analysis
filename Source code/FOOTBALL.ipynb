```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
dfm=pd.read_excel(r'C:\Users\Anay\Desktop\DATA\MAN CITY SPENDINGS(2008-20019).xlsx')
dft=pd.read_excel(r'C:\Users\Anay\Desktop\DATA\MAN CITY TROPHIES COUNT(2008-2019).xlsx')
```

```python
dft.rename({'YEARS':'YEAR'},axis='columns',inplace=True)
dft.head()
```

```python
dft=dft.drop(['TROPIES'],axis=1)
dfm=dfm.drop(['SPENDING(€) in Millions'],axis=1)
```

```python
dfm.head()
dfm.rename({'SPENDING(€) in Millions.1':'SPENDING'},axis='columns',inplace=True)
```

```python
x=dfm['YEAR']
y1=dft['TROPHIES_c']
y2=dfm['SPENDING']

fig,ax1=plt.subplots()
ax1.legend();
plt.title('MAN CITY(2008-2019)',color='lightblue',fontweight='bold')
ax1.set_xlabel('YEARS',fontweight='bold')
ax1.set_ylabel('SPENDINGS IN MILLIONS(€)',fontweight='bold')
ax1.plot(x,y2,marker="o",label='SPENDINGS IN MILLIONS(€)')
plt.legend(loc=4)
plt.xticks(x,size=8)
plt.yticks(y2,size=8)

ax2=ax1.twinx()

ax2.set_ylabel('TROPHIES',fontweight='bold')
ax2.plot(x,y1,marker="o",color='red',label='TROPHIES')
plt.yticks(y1)
plt.legend()

fig.tight_layout()
plt.show()
```
